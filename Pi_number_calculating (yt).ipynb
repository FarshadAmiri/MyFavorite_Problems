### Computing pi number using only the random function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.random(10000)
y = np.random.random(10000)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y,marker='.')
plt.xlim(0,1.2)
plt.ylim(0,1.2)
```

```python
nc = 0
for i in range(len(x)):
  if (x[i]**2+y[i]**2)<=1:
    nc += 1
print('nc = ',nc)
```

```python
nc*4/len(x)
```
